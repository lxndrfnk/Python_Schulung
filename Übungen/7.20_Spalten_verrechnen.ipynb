```python
import pandas as pd
```

```python
df = pd.read_csv("bookstore.csv")
```

```python
df.loc[:, "Added Price"] = df.loc[:, "List Price"] + df.loc[:, "Sale Price"]

# display(df.head())

df.loc[:, "Price Diff"] = df.loc[:, "List Price"] - df.loc[:, "Sale Price"]

# display(df.head())

df.loc[:, "Reduction Ratio"] = df.loc[:, "Price Diff"] / df.loc[:, "List Price"]

display(df.head())
```
