```python
import pandas as pd

df = pd.read_csv("babies.csv")
df
```

```python
df_neu = df[["bwt", "gestation"]]
df_neu
```

```python
df_neu.dropna(inplace=True)
df_neu
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# x muss in 2D-Form gebracht werden:
x = df_neu[["gestation"]]  # beachte die doppelte eckige Klammer!
y = df_neu["bwt"]

modell = LinearRegression()
modell.fit(x, y)

# Steigung (m) und Achsenabschnitt (b):
m = modell.coef_[0]
b = modell.intercept_

print(f"Gleichung: y = {m:.2f} * x + {b:.2f}")
```
